```python
import mss
import mss.tools
import datetime
```

```python
x, y = 1580, 498
for i in range(8):
    with mss.mss() as sct:
        monitor = {'top': y - 4, 'left': x - 223, 'width': 90, 'height': 9}
        y += 13
        now = datetime.datetime.now()
        output = now.strftime("%m-%d-%H-%M-%S-%f.png".format(**monitor))
        sct_img = sct.grab(monitor)
        image = str(mss.tools.to_png(sct_img.rgb, sct_img.size))
        print(image + '\n')
```
